**Conclusion: we will build a non-linear regression model to predict the logarithm of the traffic volume**

* Let's create the first predictor of the model `day_num`-- the ordinal number of the pair (day, week).
* Let's log the time series.
* Let's add squares of observation numbers. Suddenly help. The trend is non-linear...

```python
df['day_num'] = range(1, len(df) + 1)
df['log_y'] = np.log10(df['y'])
df['day_num_2'] = df['day_num']**2

# Display the first 5 rows of the resulting DataFrame
df.head()
```

```python
# Predict the logarithm, so the original series is not needed
# We do not use dates, but they were ignored when importing data
df_2 = df.drop(columns=["y"])

df_2.head()
```

```python
112/7
```

**Let's create new variables - indicators of week days**

```python
# Convert the 'date' column to a datetime object
df_2['ds'] = pd.to_datetime(df_2['ds'])

# Create binary indicator variables for each day of the week
df_2['monday'] = (df_2['ds'].dt.dayofweek == 0).astype(int)
df_2['tuesday'] = (df_2['ds'].dt.dayofweek == 1).astype(int)
df_2['wednesday'] = (df_2['ds'].dt.dayofweek == 2).astype(int)
df_2['thursday'] = (df_2['ds'].dt.dayofweek == 3).astype(int)
df_2['friday'] = (df_2['ds'].dt.dayofweek == 4).astype(int)
df_2['saturday'] = (df_2['ds'].dt.dayofweek == 5).astype(int)
df_2['sunday'] = (df_2['ds'].dt.dayofweek == 6).astype(int)
```

```python
# Create a matrix of predictors.
# This time do not remove the column corresponding to the base month from it
X = df_2.drop(columns=["log_y", "ds"])
```

```python
# Create a matrix (column) of responses.
y = df_2['log_y']
```

* Random Forest cannot predict a series with an uptrend.
* We will predict a series from the differences of the series.
* More precisely, a series of differences in the logarithms of the values of the original series.

```python
# Response - differences, still discard the pass
y_2 = y.diff(periods=1).dropna()
```

```python
# In the matrix of predictors, we discard the first row, where y_2 has NaN
X_2 = X[1:]

X_2.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state=42,
                               # число деревьев в лесу
                               n_estimators=150,
                               # функция для impurity
                               criterion='friedman_mse',
                               # то ли число слоев, то ли число расщеплений  
                               #  ?????????????????  но не число узлов
                               max_depth=10,
                               # сколько переменных использовать при построении дерева?
                               # sqrt(k)
                               max_features=10, 
                               # максимальное число конечных узлов. Сейчас - без ограничений
                               max_leaf_nodes=None,
                               # минимальное уменьшение загрязнения 
                               min_impurity_decrease=0.0, 
                                # Дерево обучается по части наблюдений (По какой???)
                               bootstrap=True, 
                               # минимальное число наблюдений в узле потомке
                               min_samples_leaf=1, 
                               # минимальное число наблюдений в узле родителе
                               min_samples_split=2,
                               # параметр используется редко, самостоятельно
                               min_weight_fraction_leaf=0.0, 
                               # объем промежуточных результатов в ходе обучения
                               verbose=1, 
                               # число используемых процессоров
                               # n_jobs=None,
                               # Вычислять out-of-bag ошибку
                               oob_score=True,
                               # использовать результаты предыдущего вызова и нарастить предыдущий лес 
                               warm_start=False)

model.fit(X_2, y_2)
```

```python
# Compare the fit and a series of differences
pred = pd.DataFrame({
     'pred': model.predict(X_2),
     'real': y_2})
pred.plot()
```

```python
# quality criterion
print("SSE = ", sum((y_2 - model.predict(X_2))**2))
```

```python
print(model.predict(X_2))
```

```python
# From fitting differences to fitting a series of logarithms
res = model.predict(X_2)
a = [df['log_y'][0]]
for i in res:
    a.append(a[-1] + i)

print(a)
```
